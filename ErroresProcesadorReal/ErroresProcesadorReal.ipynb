# Cálculo del error

```python
for x in calculado.keys():
    arrTmp = []
    for c,r in zip(calculado[x], real[x]):
        arrTmp.append(complex(abs(abs(c.real) - abs(r.real)), abs(abs(c.imag) - abs(r.imag))))
    error[x] = np.array(arrTmp)
```

## Estadísticas totales

```python
arrTotal = []
cov_mat = [[],[]]
cov_matElecR = []
cov_matElecI = []
for x in error.keys():
    arrTotal.append(error[x])
    cov_mat = np.concatenate((cov_mat, np.stack((error[x].real, error[x].imag), axis = 0)), axis=1)
    cov_matElecR.append(error[x].real)
    cov_matElecI.append(error[x].imag)
arrTotal = np.array(arrTotal)
Etotal_max = np.max(arrTotal)
Etotal_mean = np.mean(arrTotal)
Etotal_median = np.median(arrTotal)
Etotal_std = np.std(arrTotal)

Etotal_covIR = np.cov(cov_mat)
Etotal_covElectrodoR = np.cov(cov_matElecR)
Etotal_covElectrodoI = np.cov(cov_matElecI)

print("Error total:")
print("Máximo = " + str(Etotal_max))
print("Media = " + str(Etotal_mean))
print("Mediana = " + str(Etotal_median))
print("Desviación típica = " + str(Etotal_std))
print("Covarianza entre parte real e imaginaria:")
print(Etotal_covIR)

print("Covarianza entre electrodos R:")
print(Etotal_covElectrodoR)

print("Covarianza entre electrodos I:")
print(Etotal_covElectrodoI)
```
